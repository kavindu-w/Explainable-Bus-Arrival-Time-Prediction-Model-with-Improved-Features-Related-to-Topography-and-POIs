##### Running time Dataset Preperation

```python
import pandas as pd
```

```python
dataset = pd.read_csv('../../data/bus_running_time_bend_data_added.csv')
poi_dir1 = pd.read_csv('../../data/running_times_poi_final_Kandy-Digana_additional.csv')
poi_dir2 = pd.read_csv('../../data/running_times_poi_final_Digana-Kandy_additional.csv')
```

```python
dataset.head()
```

```python
poi_dir1.head()
```

```python
poi_dir2.head()
```

```python
print(poi_dir1.columns)
print(poi_dir2.columns)
```

```python
poi_dir1['direction'] = 1.0
poi_dir2['direction'] = 2.0
```

```python
poi_features = ['segment','direction','place_of_worship','place_of_education', 'hospital', 
'tourist_attractions', 'commercial_places', 'public_places','nearby_bus_stops','fuel_station','total_poi_count', 'average_poi_distance']
```

```python
#concatanate poi datasets along columns
poi_dir1 = poi_dir1[poi_features]
poi_dir2 = poi_dir2[poi_features]
poi_data = pd.concat([poi_dir1,poi_dir2])
poi_data.shape
```

```python
poi_data['segment'] = poi_data['segment'].astype('int')
poi_data.head()
```

```python
#Add elevation data to existing features
dataset_poi = pd.merge(left=dataset,right=poi_data,how='left',on=['segment','direction'])
```

```python
#Drop rows with null values
dataset_poi = dataset_poi.dropna()
```

```python
print(dataset_poi.shape)
```

```python
dataset_poi.to_csv('../../data/bus_running_time_all_features.csv')
```
